## Atingimento dos Objetivos

O objetivo deste MVP foi desenvolver um pipeline completo de dados em nuvem, abrangendo as etapas de coleta, tratamento, modelagem, carga e análise, utilizando a plataforma Databricks e a tecnologia Delta Lake. Esse objetivo foi atingido de forma satisfatória.

Foi possível estruturar um pipeline organizado em camadas (Bronze, Silver e Gold), assegurando rastreabilidade, qualidade e consistência dos dados, além da construção de um Data Warehouse em modelo estrela, adequado para análises analíticas no contexto de manutenção preditiva. As análises realizadas permitiram responder às principais perguntas de negócio, evidenciando padrões operacionais e fatores associados à ocorrência de falhas.

---

## Dificuldades Encontradas

As principais dificuldades estiveram relacionadas às limitações do Databricks Community Edition, especialmente no que diz respeito à escrita em diretórios do DBFS e à necessidade de adaptação entre DataFrames PySpark e tabelas acessíveis via SQL.

Esses desafios exigiram ajustes na estratégia de persistência e padronização dos dados, contribuindo para um melhor entendimento de conceitos como governança, reexecução de pipelines e boas práticas de engenharia de dados em ambientes de nuvem.

---

## Avaliação da Qualidade da Solução

A solução desenvolvida apresentou alto nível de qualidade dos dados. As verificações realizadas confirmaram ausência de valores nulos, domínios inconsistentes ou problemas de integridade referencial no Data Warehouse.

A modelagem dimensional adotada mostrou-se adequada para suportar análises analíticas e segmentações relevantes, facilitando a exploração dos dados e a interpretação dos resultados obtidos.

---

## Trabalhos Futuros

Como evolução do MVP, destacam-se as seguintes possibilidades:

- Automatização do pipeline por meio de mecanismos de orquestração, como Databricks Jobs ou Apache Airflow;
- Integração com fontes de dados em tempo real ou APIs industriais;
- Desenvolvimento de dashboards interativos para acompanhamento contínuo dos indicadores de manutenção;
- Expansão do modelo dimensional com novas dimensões analíticas;
- Integração de modelos de machine learning ao pipeline para predição de falhas.

---

## Considerações Finais

O desenvolvimento deste MVP permitiu consolidar, de forma prática, conceitos fundamentais de engenharia de dados, pipelines em nuvem e modelagem analítica. A experiência reforçou a importância de arquiteturas bem estruturadas para garantir qualidade, escalabilidade e suporte efetivo à tomada de decisão baseada em dados.
